```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
admissions= pd.read_csv("../data/mimic-iii-clinical-database-demo-1.4/ADMISSIONS.csv")
admissions.head(5)
admissions.info()

```

```python
print(admissions.describe(include="all"))
print(admissions.isnull().sum())
```

VISUALISE DATA


```python
survive= admissions['hospital_expire_flag'].value_counts()
print(survive)

plt.figure(figsize=(6,4))
sns.set(style="whitegrid")

ax= sns.countplot(x='hospital_expire_flag', palette="Set1", data=admissions)
for container in ax.containers:
    ax.bar_label(container)
plt.title("0= survived and 1 = died")
plt.xticks(rotation=45)
plt.ylabel("count")
plt.show()
```

In Seaborn plots, hue is used to split your data by a categorical variable and show it using different colors.

Think of it as “color by category.”

ADMISSION TYPE AND SURVIVAL

```python
ax = sns.countplot(
    x='admission_type',
    hue='hospital_expire_flag',
    data=admissions,
    palette='coolwarm',
    order=admissions['admission_type'].value_counts().index
)

for container in ax.containers:
    ax.bar_label(container)

plt.title("Admission Type vs Hospital Outcome")
plt.xlabel("Admission Type")
plt.ylabel("Count")
plt.show()
```

The histogram above shows the distribution of hospital stay lengths (in days) for patients in the dataset. Each bar represents the number of patients who stayed for a given duration. The bars are split by survival outcome (hospital_expire_flag)

Length of stay od patients

```python
admissions['admittime']=pd.to_datetime(admissions['admittime'])
admissions['dischtime']=pd.to_datetime(admissions['dischtime'])

admissions['len_of_stay']= (admissions['dischtime']-admissions['admittime']).dt.days
print(admissions['len_of_stay'])

plt.figure(figsize=(6,4))
sns.histplot(admissions['len_of_stay'],bins=20,kde=True)
plt.title("Time of stay distribution")
plt.xlabel("Days")
plt.show()
```

ADMISSION TIME AND SURVIVAL

```python
plt.figure(figsize=(8,5))
ax=sns.histplot(
    data=admissions,
    x='len_of_stay',
    hue='hospital_expire_flag',
    bins=20,
    palette='Set1',
    multiple='stack',  
    kde=False
)
for container in ax.containers:
    ax.bar_label(container)
plt.title("Length of Stay Distribution (0=survived 1= died)")
plt.xlabel("Length of Stay")
plt.ylabel("Number of Patients")
plt.show()
```

CREATING FUNCTION FOR CATEGORICAL VS TARGET VALUE VISUALS

```python
def compare_data(df, feature, target='hospital_expire_flag', palette='Set2'):

    plt.figure(figsize=(8,5))
    
    # Countplot
    ax = sns.countplot(
        x=feature,
        hue=target,
        data=df,
        order=df[feature].value_counts().index,
        palette=palette
    )
    
    # Add exact counts on top of bars
    for container in ax.containers:
        ax.bar_label(container)
    
    plt.title(f"{feature} vs {target}")
    plt.xlabel(feature)
    plt.ylabel("Number of Patients")
    plt.xticks(rotation=45)
    plt.show()

categorical_features = ['admission_type', 'insurance', 'marital_status']

for feature in categorical_features:
    compare_data(admissions, feature)
```
